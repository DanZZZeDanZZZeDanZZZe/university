```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import random as rnd
```

## Инициализируем
- число испытаний
- положение экранов
- начальное положение
- вероятность шага вправо

```python
n = 100000
a, b = 0, 10
x0 = 5
p = 0.5
```

## Функция моделирования блуждания

```python
def inc(x):
    x += 1
    return x


def dec(x):
    x -= 1
    return x


def simulate_wanderings(p, x0, a, b):
    x = x0
    time = 0
    while ((x != a) and (x != b)):
        r = rnd.random()
        x = inc(x) if r < p else dec(x)
        time = inc(time)
    return time
```

## Вычисления

```python
results = [simulate_wanderings(p, x0, a, b) for _ in range(n)]
average_time = np.mean(results)
dispersion = np.var(results)
```

## Результаты

```python
division_value = 10

plt.figure(figsize=(12, 20))
frequency_histogram = plt.subplot(211)
static_histogram = plt.subplot(212)

frequency_histogram.set_title('Гистограмма частот')
static_histogram.set_title('Статистическая функция распределения')

frequency_histogram.xaxis.set_major_locator(
    ticker.MultipleLocator(division_value))
static_histogram.xaxis.set_major_locator(
    ticker.MultipleLocator(division_value))

frequency_histogram.grid(True)
static_histogram.grid(True)

frequency_histogram.hist(results, bins=20, density='True', histtype='bar', facecolor='orange')
static_histogram.hist(results, bins=20, density='True', histtype='bar', cumulative='True', facecolor='green')
```
